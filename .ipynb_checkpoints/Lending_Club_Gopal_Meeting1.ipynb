```python
# check the number of Nan or Null for each column
#As per our objective, Filtering only the columns required
current_df = df.filter(col('loan_status').isin(['Late (31-120 days)','Charged Off','Late (16-30 days)','Current']))
#Renaming the Late values into a single late
current_df = current_df.withColumn("loan_status", \
              when(current_df["loan_status"].isin(['Late (31-120 days)','Late (16-30 days)']),'Late').otherwise(current_df["loan_status"]))
#Our dataset with only the desired columns values (Late,charged off,current)
print(current_df.groupby(current_df.loan_status).count().show())
size_shape(current_df)
```

```python
#Contains the nas present in all the columns in the dataset. We are planning to remove those coluomns with more than 50%Nas
checkna_pd = current_df.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in current_df.columns]).toPandas()
checkna_pd

# current_df.select([c for c in checkna_pd.columns])
```

```python
#https://stackoverflow.com/questions/51322445/how-to-drop-all-columns-with-null-values-in-a-pyspark-dataframe - include in our references

#Creating a dataframe which has the count of Nas of each column
checkna = current_df.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in current_df.columns])
#Selecting the columns with NA values not greater than 50%
No_of_rows = current_df.count()
checkna_greater_than_50 = checkna.select([when(fn.col(c)<(No_of_rows*0.50),c).alias(c) for c in checkna.columns])


def drop_null_columns(df):
    """
    This function drops all columns which contain null values.
    :param df: A PySpark DataFrame
    """
    null_counts = df.select([fn.count(fn.when(fn.col(c).isNull()|isnan(fn.col(c)), c)).alias(c) for c in df.columns]).collect()[0].asDict()
    to_drop = [k for k, v in null_counts.items() if v > 0]
    df = df.drop(*to_drop)
    return df

# Using the function to find the list of columns with less than 50%Na values
final_cols = drop_null_columns(checkna_greater_than_50)
current_df1 = current_df.select([fn.col(c) for c in final_cols.columns])
size_shape(current_df)
size_shape(current_df1)
#selecting our dataframe with the final cols
```

```python
#Now 1st let us find the type aka.. categorical or just plain number type among numerical
# converting the double type to int type

from pyspark.sql.types import *
part4 = part4.withColumn('pct_tl_nvr_dlq', part4['pct_tl_nvr_dlq'].cast(IntegerType()))
part4 = part4.withColumn('percent_bc_gt_75', part4['percent_bc_gt_75'].cast(IntegerType()))
#So, only 4 are the continous numerical and others are just categorical(Also checked w.r.t data definitions)
numerical_cols = ["tot_hi_cred_lim","total_bal_ex_mort","total_bc_limit","total_il_high_credit_limit"]
string_cols = ['disbursement_method','debt_settlement_flag','hardship_flag']
#So below, Other than the last 4 every other column is categorical(Also checked w.r.t data descriptions). I'll append this to our already existing categorical cols
categorical_cols = [c for c in part4.columns if c not in numerical_cols]
categorical_numerical = [c for c in categorical_cols if c not in string_cols]

#So 3 types of Cols - numerical,categorical_numerical & String cols
```

```python
from pyspark.ml.feature import Imputer
from pyspark.ml import Pipeline


# Now let us start filling out the NAs

#Now then, for the numerical cols, let us fill the NAs with mean & Winsorize them all
mean_numerical = part4.select([fn.avg(cols).alias(cols) for cols in numerical_cols]).first().asDict()
part4 = part4.na.fill(mean_numerical)

#Now for categorical numerical coles.. we fill by median
part4_dup = part4
imputer = Imputer(inputCols=categorical_numerical,outputCols=categorical_numerical,strategy = 'median')
part4 = imputer.fit(part4).transform(part4)

#Now for the string Cols - Since those are categorical, I will replace it by the value 'unknown'. 
part4 = part4.fillna('unknown',subset = string_cols)


part4.select([fn.count(fn.when(fn.isnan(c),c)).alias(c) for c in part4.columns]).toPandas()
```

```python
#LEt us winsorize the numerical data
# empty dictionary d
d = {}
# Fill in the entries one by one
for col in numerical_cols:
      d[col] = part4.approxQuantile(col,[0.01,0.99],0.25)
      print(col+" done")

part4.select([c for c in numerical_cols]).describe().show()

for c in numerical_cols:
    part4 = part4.withColumn(c,fn.when(fn.col(c)<d[c][0],d[c][0]).when(fn.col(c)>d[c][1],d[c][1]).otherwise(fn.col(c)))
    print(c + ' done')
    
#Guess there is no winsorize necessary
part4.select([c for c in numerical_cols]).describe().show()   
```

```python
cleaned_dataframe_part4 = part4
```

```python
all_cols = []
# Now that cleaning is done ; Let us start with the pipeline
# Let us convert all to numeric
for c in string_cols:
    part4 = Pipeline(stages = [feature.StringIndexer(inputCol = c,outputCol = c+'index'),
                              feature.OneHotEncoder(inputCol = c+'index',outputCol = c+'stringcoded')]).fit(part4).transform(part4)
    all_cols.append(c+'stringcoded')
for c in categorical_numerical:
    print(c)
    part4 = Pipeline(stages = [feature.OneHotEncoder(inputCol = c,outputCol = c+'catcoded')]).fit(part4).transform(part4)
    all_cols.append(c+'catcoded')
```
